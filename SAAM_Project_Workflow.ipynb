# SAAM Project Workflow

This notebook gives a simple entry point to the project.
It is meant to stay readable: we run the scripts in order and then inspect the main output files.

```python
from pathlib import Path
import subprocess
import sys
import pandas as pd

BASE_DIR = Path().resolve()
SRC_DIR = BASE_DIR / "src"
PROCESSED_DIR = BASE_DIR / "data" / "processed"

assert SRC_DIR.exists(), f"Missing folder: {SRC_DIR}"
assert (BASE_DIR / "data" / "Raw").exists(), "Missing Raw data"
```

```python
def run_script(script_name: str) -> None:
    script_path = SRC_DIR / script_name
    print(f"Running {script_name}...")
    subprocess.run([sys.executable, str(script_path)], cwd=str(BASE_DIR), check=True)
    print(f"Finished {script_name}.\n")
```

## Full run order

Run the scripts in this order if you want to rebuild the whole project.

```python
FULL_WORKFLOW = [
    "p1_1_data_cleaning.py",
    "p2_1_investment_set.py",
    "p2_2_minimum_variance_portfolio.py",
    "p2_3_value_weighted_portfolio.py",
    "p2_4_minvar_vs_valuweighted_comparison.py"
    "p3_1_carbon_footprint.py",
    "p3_2_minimum_variance_carbon.py",
    "p3_3_tracking_error_carbon.py",
    "p3_4_carbon_comparison.py",
    "p4_1_net_zero.py",
    "p4_2_passive_comparison.py",
]

FULL_WORKFLOW
```

## Launch the whole workflow

Running the next cell rebuilds all project outputs from the raw data.

```python
for script in FULL_WORKFLOW:
    run_script(script)
```
